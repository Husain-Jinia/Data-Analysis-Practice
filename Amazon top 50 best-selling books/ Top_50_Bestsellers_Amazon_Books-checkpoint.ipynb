**DATA WRANGLING**

categorizing data for effective analysis

```python
grp = df.groupby('Genre') 
grp1= df.groupby('Author')
x= grp['Price'].agg(np.sum)
y= grp['Rating'].agg(np.mean)
z= grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
y1 = grp1['Rating'].agg(np.mean)
x1= grp1['Price'].agg(np.sum)
z1= grp1['Reviews'].agg(np.mean)
print(y1)
print(x1)
print(z1)
```

```python
grp3 = df.groupby("Year")
x2=grp3['Price'].agg(np.sum)
print(x2)
```
