```python
#changing the format of the date and time column so that it is easier to read

df['datetime_utc'] = pd.to_datetime(df['datetime_utc'])
df.head(10)
```

```python
#Keeping the necessary required column and discarding the rest for ease of reading the data

new_df=pd.DataFrame(list(df[' _tempm']),index=df.datetime_utc, columns=['temperature'])
new_df
```

```python
#Getting the mean for the temperature throughout the day by calculating the mean of the temperature given on the different time of the particular day

new_df=new_df.resample('D').mean()
new_df
```

```python
#Getting the mean for the temperature throughout the month by calculating the mean of the temperature throughout the days of that month

month_df=new_df.resample('M').mean()
month_df
```

```python
#Getting the mean for the temperature throughout the year by calculating the mean of the temperature throughout the months of that year

year_df=new_df.resample('Y').mean()
year_df
```

```python
#plotting graph to check the change in temperature from 1995 to 2018
plt.figure(figsize=(25,10))
plt.plot(range(1996,2018,1),year_df.temperature)
plt.xlabel('Year')
plt.ylabel('Temperature in Celseius')
```
