```python
popularity = [0,0,0,0]
for i in df["User Rating"]:
    if i >= 4.8:
        popularity[0] += 1
    elif i >= 4.5:
        popularity[1] += 1
    elif i >= 4.0:
        popularity[2] += 1
    else:
        popularity[3] += 1
```

```python
x = ['Extremely Popular(4.8, 4.9)','Very Popular(4.4 - 4.7)','Fairly Popular(4.0 - 4.3)','Popular (- 3.9)']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x,popularity)
plt.title('Popularity of books')
plt.xticks(rotation=90)
plt.show()
```

```python
top_rating = df[df['User Rating'] == df['User Rating'].max()]
top_rating.set_index("Year").sort_values(by="Year", ascending=True)
top_rating
```

```python
tr = top_rating.groupby("Author")
t_r = tr["User Rating"].agg(np.mean)
print(t_r)
```

```python
grp = df.groupby('Genre') 
y= grp['User Rating'].agg(np.mean)
z= grp['Reviews'].agg(np.mean)
print(y)
print(z)
```

```python
plt.figure(figsize=(15,6)) #plotting the graph to check wich genre got better rating
plt.plot(y, 'ro', color='green')
plt.title("Genre Ratings")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6)) #plotting the graph to check which genre got the most reviews
plt.plot(z, 'ro',color="blue")
plt.title("Genre Reviews")
plt.xticks(rotation=90)
plt.show()
```
